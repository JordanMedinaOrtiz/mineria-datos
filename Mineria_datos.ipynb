**Minería de Datos (Data Mining)**

La minería de datos es el proceso de analizar grandes volúmenes de datos para descubrir patrones, anomalías y correlaciones ocultas, transformando información cruda en conocimiento valioso para:
- Toma de decisiones estratégicas
- Predicción de tendencias
- Optimización de procesos en diversas industrias

**En términos prácticos:** Es inteligencia aplicada a la información. Consiste en usar programas informáticos avanzados para revisar montañas de datos y encontrar "pistas" que un ser humano jamás vería por su cuenta. Su objetivo no es solo saber qué pasó ayer, sino **predecir qué pasará mañana** (como qué comprará un cliente o si una transacción es fraudulenta) para tomar decisiones más inteligentes y rentables.

**Ciencia de Datos (Data Science)**

Es un campo multidisciplinario que usa métodos científicos, procesos, algoritmos y sistemas para extraer conocimiento e insights valiosos de grandes volúmenes de datos (Big Data), combinando estadística, matemáticas, inteligencia artificial (IA) e informática para analizar, predecir y tomar decisiones informadas.

**En términos prácticos:** Es el ecosistema completo del manejo de información. No es solo una técnica, es una disciplina híbrida que mezcla tres mundos:

- **Matemáticas y Estadística:** Para modelar la realidad
- **Programación y Tecnología:** Para procesar volúmenes masivos (Big Data)
- **Conocimiento del Negocio:** Para que los resultados tengan sentido útil

Su misión abarca todo el ciclo de vida del dato: desde que nace (captura), se limpia, se analiza, hasta que se convierte en un producto o una decisión ejecutiva.

####Análisis Estadístico Básico con NumPy

Media (Mean)

```python
# La media es el promedio aritmético: suma de valores / cantidad
# np.mean() calcula esto en un array de forma eficiente
data = np.array([1, 2, 3, 4, 5])
media = np.mean(data)
print(media)  # Resultado: 3.0
```

Promedio (Average) y Promedio Ponderado

```python
# np.average() sin pesos es igual a mean()
# Con weights= podemos darle más importancia a ciertos valores
# Útil cuando algunos datos son más relevantes que otros
data = np.array([1, 2, 3, 4, 5])
weights = np.array([1, 2, 3, 4, 5])  # El 5 pesa 5 veces más que el 1
promedio = np.average(data)
promedio_ponderado = np.average(data, weights=weights)

print(promedio)            # Promedio simple: 3.0
print(promedio_ponderado)  # Promedio ponderado: 3.67 (más peso al final)
```

Mediana

```python
# La mediana es el valor del medio cuando ordenas los datos
# Menos sensible a valores extremos (outliers) que la media
mediana = np.median(data)

print(mediana)  # Resultado: 3.0
```

Varianza y Desviación Estándar

```python
# Varianza: promedio de las diferencias al cuadrado respecto a la media
# Mide qué tan dispersos están los datos
# Desviación estándar: raíz cuadrada de la varianza (mismas unidades que los datos)
varianza = np.var(data)
desviacion = np.std(data)

print(varianza)   # Varianza: 2.0
print(desviacion) # Desv. estándar: 1.414...
```

Generación de Datos Aleatorios

```python
# random.normal() genera datos con distribución normal (campana de Gauss)
# (media=0, desv_std=1, tamaño=1000) - distribución estándar
muestra = np.random.normal(0, 1, 1000)
print(np.mean(muestra), np.std(muestra))  # Debería ser ~0 y ~1
```

Correlación

```python
# corrcoef() calcula la matriz de correlación entre variables
# Valores cercanos a 1 = correlación positiva fuerte
# Valores cercanos a -1 = correlación negativa fuerte
# Valores cercanos a 0 = sin correlación
corr = np.corrcoef([1,2,3], [1,5,7])
print(corr)  # Matriz 2x2 con correlaciones
```

```python
# Creamos un resumen estadístico completo de la variable edad
# Incluye métricas de tendencia central, dispersión y posición
summary_age = {
    'n': int(edad.size),                           # Cantidad de observaciones
    'min': float(np.min(edad)),                    # Valor mínimo
    'p25': float(np.percentile(edad, 25)),         # Primer cuartil (25%)
    'mediana': float(np.median(edad)),             # Mediana (50%)
    'media': float(np.mean(edad)),                 # Media aritmética
    'p75': float(np.percentile(edad, 75)),         # Tercer cuartil (75%)
    'max': float(np.max(edad)),                    # Valor máximo
    'rango(ptp)': float(np.max(edad) - np.min(edad)),  # Rango (max - min), np.ptp está deprecado
    'var_muestral': float(np.var(edad, ddof=1)),  # Varianza muestral (ddof=1 corrige sesgo)
    'std_muestral': float(np.std(edad, ddof=1)),  # Desviación estándar muestral
    'cv': float(np.std(edad, ddof=1)/np.mean(edad)),  # Coeficiente de variación (dispersión relativa)
}

summary_age
```

```python
# Calculamos frecuencias absolutas de las variables booleanas
# count_nonzero() cuenta cuántos True hay (respuestas "Sí")
freq_python = {'Si': int(np.count_nonzero(prog_py)), 'No': int(edad.size - np.count_nonzero(prog_py))}
freq_colab  = {'Si': int(np.count_nonzero(usa_colab)), 'No': int(edad.size - np.count_nonzero(usa_colab))}
freq_gh     = {'Si': int(np.count_nonzero(cuenta_gh)), 'No': int(edad.size - np.count_nonzero(cuenta_gh))}

print(f'Frecuencia de utilizar Python: {freq_python}')
print(f'Frecuencia de utilizar Google Colab: {freq_colab}')
print(f'Frecuencia de tener cuenta de GitHub: {freq_gh}')
```

```python
# Función para crear tablas de contingencia 2x2 entre dos variables booleanas
# Útil para ver cómo se cruzan dos características (ej: Python + Colab)
def tabla2x2(a, b):
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a & b)    # Ambos True (Sí y Sí)
    t[1,0] = np.count_nonzero(a & ~b)   # a=True, b=False (Sí y No)
    t[0,1] = np.count_nonzero(~a & b)   # a=False, b=True (No y Sí)
    t[0,0] = np.count_nonzero(~a & ~b)  # Ambos False (No y No)
    return t

# Generamos las tres combinaciones que nos interesan
tab_py_colab = tabla2x2(prog_py, usa_colab)
tab_py_gh    = tabla2x2(prog_py, cuenta_gh)
tab_colab_gh = tabla2x2(usa_colab, cuenta_gh)
```

```python
# Mostramos las tablas de contingencia para analizar relaciones
print(f'Tabla de contingencia de Python y Google Colab:\n{tab_py_colab}')
print(f'Tabla de contingencia de Python y GitHub:\n{tab_py_gh}')
print(f'Tabla de contingencia de Google Colab y GitHub:\n{tab_colab_gh}')
```

```python
# ====================================================
# GRÁFICO 3: Barras para frecuencias de Sí/No
# ====================================================
# Tres subplots uno al lado del otro, compartiendo eje Y
fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)

# Iteramos sobre cada variable booleana
for ax, col, title, color in zip(
    axes,
    ['prog_python','usa_colab','cuenta_github'],
    ['Programó en Python','Usó Google Colab','Cuenta de GitHub'],
    [accent, accent2, accent3]
):
    # Contamos Sí/No y rellenamos con 0 si alguno falta
    counts = df[col].value_counts().reindex([True, False]).fillna(0).astype(int)
    labels = ['Sí','No']
    # Sí con color específico, No con gris neutral
    ax.bar(labels, counts, color=[color, neutral], edgecolor='white')
    ax.set_title(title)
    ax.set_ylim(0, max(counts.max(), 1)+2)
    # Etiquetas de valores arriba de cada barra
    for i, v in enumerate(counts):
        ax.text(i, v+0.1, str(v), ha='center', va='bottom', fontsize=9)

axes[0].set_ylabel('Número de estudiantes')
plt.tight_layout()
plt.show()
fig.savefig('03_barras_frecuencias.png', dpi=180)
```

```python
# =======================================
# GRÁFICO 4: Mapas de calor para tablas 2x2
# =======================================
# Visualizamos las relaciones entre pares de variables
pairs = [
    ('prog_python','usa_colab','Python vs Colab'),
    ('prog_python','cuenta_github','Python vs GitHub'),
    ('usa_colab','cuenta_github','Colab vs GitHub')
]

for idx, (a,b,titulo) in enumerate(pairs, start=1):
    # Convertimos a arrays booleanos
    a_bool = df[a].to_numpy(dtype=bool)
    b_bool = df[b].to_numpy(dtype=bool)

    # Construimos tabla de contingencia con operaciones lógicas
    t = np.zeros((2,2), dtype=int)
    t[1,1] = np.count_nonzero(a_bool & b_bool)   # Ambos Sí
    t[1,0] = np.count_nonzero(a_bool & ~b_bool)  # a=Sí, b=No
    t[0,1] = np.count_nonzero(~a_bool & b_bool)  # a=No, b=Sí
    t[0,0] = np.count_nonzero(~a_bool & ~b_bool) # Ambos No

    # Creamos el heatmap
    fig, ax = plt.subplots(figsize=(5,4))
    im = ax.imshow(t, cmap='Blues')  # Colores azules con intensidad proporcional
    ax.set_title(f'Tabla 2×2: {titulo}')
    ax.set_xticks([0,1]); ax.set_xticklabels(['No','Sí'])
    ax.set_yticks([0,1]); ax.set_yticklabels(['No','Sí'])
    ax.set_xlabel(b); ax.set_ylabel(a)

    # Anotamos cada celda con su valor
    for i in range(2):
        for j in range(2):
            ax.text(j, i, t[i,j], ha='center', va='center',
                    color='black', fontsize=12, fontweight='bold')

    cbar = fig.colorbar(im, ax=ax)
    cbar.ax.set_ylabel('Conteo', rotation=270, labelpad=12)
    plt.tight_layout()
    plt.show()
    fig.savefig(f'04_heatmap_{idx}.png', dpi=180)
```

```python
# ======================================================
# GRÁFICO 5: Edad promedio por grupo (Sí/No)
# ======================================================
# Analizamos si hay diferencia de edad entre quienes usan vs no usan cada herramienta
fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)

for ax, col, title, color in zip(
    axes,
    ['prog_python','usa_colab','cuenta_github'],
    ['Edad promedio por Python','Edad promedio por Colab','Edad promedio por GitHub'],
    [accent, accent2, accent3]
):
    # Agrupamos por True/False y calculamos media de edad
    means = df.groupby(col)['edad'].mean().reindex([True, False])
    labels = ['Sí','No']
    bars = ax.bar(labels, means, color=[color, neutral], edgecolor='white')

    # Etiquetas con 2 decimales
    for i, v in enumerate(means):
        ax.text(i, v+0.05, f"{v:.2f}", ha='center', va='bottom', fontsize=9)

    ax.set_title(title)
    ax.set_xlabel('Grupo')

axes[0].set_ylabel('Edad promedio (años)')
plt.tight_layout()
plt.show()
fig.savefig('05_barras_edad_promedio.png', dpi=180)
```

Primer intento de modelar los años de escolaridad y el ingreso estimado.   
Utilizar el primer punto y el útlimo para trazar una recta:   
$$ m = \frac{y_{2}-y_{1}}{x_{2}-x_{1}} $$
$$ b = y_{1} - m * x_{1} $$

```python
x1 = anios_escolaridad[0]
y1 = ingreso_mensual_estimado_mx[0]
x2 = anios_escolaridad[-1]
y2 = ingreso_mensual_estimado_mx[-1]

m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

print(f"La ecuación de la recta es: y = {m:.2f}x + {b:.2f}")
```

```python
import matplotlib.pyplot as plt

# Calcular los valores y de la recta utilizando m y b
linea_ingreso_estimado_new_plot = m * anios_escolaridad + b

plt.figure(figsize=(10, 6))
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx, label='Datos Originales')
plt.plot(anios_escolaridad, linea_ingreso_estimado_new_plot, color='red',
         label=f'Recta de Regresión (y = {m:.2f}x + {b:.2f})')
plt.title('Años de Escolaridad vs. Ingreso Mensual Estimado con Línea de Regresión')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.legend()
plt.show()
```

Valores predichos  $(y_{pred})$: Son los ingresos mensuales que nuestra recta de regresión predice para cada año de escolaridad.

Errores (residuales): Es la diferencia entre el ingreso mensual real (ingreso_mensual_estimado_mx) y el ingreso mensual predicho (y_pred). Un error positivo significa que la predicción fue menor que el valor real, y un error negativo indica que la predicción fue mayor.

Errores al cuadrado: Hemos elevado al cuadrado cada uno de los errores. Esto se hace para eliminar los signos negativos (de modo que los errores positivos y negativos no se cancelen entre sí) y para penalizar más los errores grandes.

Suma de errores cuadrados: Es la suma de todos los errores al cuadrado. Este valor nos da una medida total de la 'distancia' de nuestros puntos a la línea.

Error Cuadrático Medio (MSE): Es la suma de los errores cuadrados dividida por el número total de puntos de datos. El MSE es una medida promedio de la magnitud de los errores. Un MSE más bajo indica que el modelo se ajusta mejor a los datos.

En este caso, el Error Cuadrático Medio (MSE) es de aproximadamente 13,196,667.31. Este número nos da una idea del rendimiento general de la recta que trazamos, mostrando la magnitud promedio de los errores al cuadrado entre los valores observados y los valores predichos por la línea.

Fórmula del MSE (Error Cuadrático Medio) del Inglés Mean Squared Error
$$ MSE = \frac{\Sigma_{i=1}^{n}(Y_{i}-\hat{Y}_{i})^2}{n} $$
Donde:   
$MSE =$ error cuadrático medio.   
$n =$ número de datos.   
$Y_{i} =$ valores observados.   
$\hat{Y}_{i} =$ valores predichos.   

```python
# 1. Calcular los valores predichos (y_pred) usando la ecuación de la recta
y_pred = m * anios_escolaridad + b
y_pred
```

```python
# 2. Calcular los errores (residuales): la diferencia entre los valores reales y los predichos
errores = ingreso_mensual_estimado_mx - y_pred
errores
```

```python
# 3. Elevar al cuadrado cada error
errores_cuadrados = errores**2
errores_cuadrados
```

```python
# 4. Sumar los errores cuadrados
suma_errores_cuadrados = np.sum(errores_cuadrados)
suma_errores_cuadrados
```

```python
# 5. Dividir la suma de los errores cuadrados por el número de puntos para obtener el MSE
mse = suma_errores_cuadrados / len(anios_escolaridad)
mse
```

Ahora comparemos el MSE de una regresión lineal.   
Por ahora no es importante el detalle de la regresión lineal.   
Lo importante es comparar el MSE.

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Cambia la forma de anios_escolaridad a 2D array como es requerido por sklearn
X = anios_escolaridad.reshape(-1, 1)
y = ingreso_mensual_estimado_mx

X
```

```python
# Crear y ajustar el modelo de regresión lineal
modelo_regresion = LinearRegression()
modelo_regresion.fit(X, y)

# Obtener los coeficientes (m) e intercepto (b) del nuevo modelo
m_sklearn = modelo_regresion.coef_[0]
b_sklearn = modelo_regresion.intercept_

print(f"Coefficients (m) from scikit-learn: {m_sklearn:.2f}")
print(f"Intercept (b) from scikit-learn: {b_sklearn:.2f}")
```

El nuevo modelo de recta obtenido con regrsión lineal es:   
$$ y = 1914.55 * x + 9731.66 $$

```python
# ===================================================================
# Ejercicio 1: Gráfica comparativa de ambas regresiones
# Mostramos datos originales + recta manual + recta sklearn
# ===================================================================

# Generamos las predicciones del modelo entrenado de sklearn
y_pred_sklearn = modelo_regresion.predict(X)

plt.figure(figsize=(10, 6))
# Datos originales como scatter plot
plt.scatter(anios_escolaridad, ingreso_mensual_estimado_mx,
            color='black', s=80, zorder=5, label='Datos Originales')
# Recta manual (obtenida con el primer y último punto)
plt.plot(anios_escolaridad, y_pred, color='red', linewidth=2,
         label=f'Recta manual (y = {m:.2f}x + {b:.2f})')
# Recta de sklearn (mínimos cuadrados ordinarios)
plt.plot(anios_escolaridad, y_pred_sklearn, color='blue', linewidth=2, linestyle='--',
         label=f'Sklearn OLS (y = {m_sklearn:.2f}x + {b_sklearn:.2f})')

plt.title('Comparativa: Regresión Manual vs Regresión Sklearn')
plt.xlabel('Años de Escolaridad')
plt.ylabel('Ingreso Mensual Estimado (MXN)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ===================================================================
# Ejercicio 2 y 3: Cálculo del MSE de sklearn y comparativa
# ===================================================================

# Calculamos el MSE de sklearn con la misma fórmula que usamos para el manual
errores_sklearn = ingreso_mensual_estimado_mx - y_pred_sklearn
mse_sklearn = np.mean(errores_sklearn ** 2)

# Comparativa cuantitativa entre ambos modelos
diferencia_mse = mse - mse_sklearn
porcentaje_mejora = (diferencia_mse / mse) * 100

print("=" * 60)
print("COMPARATIVA DE MODELOS - MSE")
print("=" * 60)
print(f"MSE Recta Manual (2 puntos):    {mse:>15,.2f}")
print(f"MSE Regresión Sklearn (OLS):    {mse_sklearn:>15,.2f}")
print(f"Diferencia (Manual - Sklearn):  {diferencia_mse:>15,.2f}")
print(f"Mejora porcentual de Sklearn:   {porcentaje_mejora:>14.2f}%")
print("-" * 60)

if mse_sklearn < mse:
    print("\nLa recta de Sklearn tiene MENOR error.")
    print("Razón: sklearn aplica Mínimos Cuadrados Ordinarios (OLS), que")
    print("minimiza la suma total de errores cuadrados considerando TODOS")
    print("los puntos. La recta manual solo pasa por el primer y último")
    print("punto, ignorando los intermedios y generando mayor desviación.")
elif mse_sklearn > mse:
    print("\nLa recta manual tiene MENOR error.")
else:
    print("\nAmbos modelos tienen el MISMO error.")
print("=" * 60)
```

**¿Qué es la probabilidad condicional?**

La **probabilidad condicional** mide la probabilidad de que ocurra un evento **A**, sabiendo que ya ocurrió un evento **B**.
$$ {P(A|B)} $$
Y su fórmula es:
$$ P(A|B) = \frac{P(A \cap B)}{P(B)}$$

donde:

${P(A \cap B)}$: probabilidad de que ocurran A y B a la vez.     
${P(B)}$: probabilidad de que ocurra B.     
⚠️ Importante: solo tiene sentido si ${P(B) > 0}$.

**2. Ejemplo básico**     
**Ejemplo:**
En una bolsa hay 5 bolas rojas y 3 verdes. Tomamos una bola al azar.
Sea:
*   A = “la bola es roja”
*   B = “la bola es verde o roja” (o sea, cualquier bola…)     

Encuentra ${P(A|B)}$     
Como B es todo el espacio, entonces ${P(A|B) = P(A)}$.

**3. Ejemplo más interesante**
**Ejemplo:**
En un salón:


*   12 estudiantes hablan inglés
*   8 hablan francés
*   5 hablan ambos idiomas
*   Hay 20 estudiantes en total

Encuentra:

a) **P(inglés | francés)**     

**Interpretación:**    
Entre quienes hablan francés, el 62.5% también habla inglés.

**Ejercicios de probabilidad condicional**     
La población de adultos en un pequeño pueblo que han satisfecho los requisitos para graduarse en la escuela. Se deben clasificar de acuerdo con su sexo y a si trabajan o no actualmente.

| | Empleado | Desempleado | Total |
| :--- | :---: | :---: | :---: |
| **Hombre** | 460 | 40 | 500 |
| **Mujer** | 140 | 260 | 400 |
| **Total** | **600** | **300** | **900** |

Se seleccionará al azar a uno de esto individuos para que realicen un viaje a través de todo el país, con la intención de promocionar las ventajas que se derivan del establecimiento de nuevas industrias en los pequeños poblados. El interés se centra en los siguientes eventos:    
H: se escoge un hombre.     
E: el elegido tiene empleo.     

Calcule $ P(H|E) $    
Paso 1: Identificar la Condición (El Denominador)     
La condición es $E$ ("el elegido tiene empleo"). Esto significa que nuestro universo se reduce solo a la columna de "Empleado". Ignoramos a los desempleados por completo.     

*   Mirando la columna "Empleado", el Total es: 600.

Paso 2: Identificar el evento deseado dentro de esa condición (El Numerador)     
Dentro de ese grupo de 600 personas empleadas, ¿cuántos son Hombres ($H$)?

*   Mirando la fila "Hombre" dentro de la columna "Empleado", el número es: 460.

Paso 3: Calcular     
Sustituimos los valores en la fórmula:$$P(H|E) = \frac{460}{600}$$
Paso 4: Simplificar y Resultado     
Podemos simplificar la fracción:     
1. Eliminamos los ceros: $\frac{46}{60}$
2. Sacamos mitad: $\frac{23}{30}$

Resultado final:
La probabilidad es de $\frac{23}{30}$ o $76.67\%$.

      

*   $P(A) = 0.60$ (Probabilidad de que sea de la máquina A)
*   $P(B) = 0.40$ (Probabilidad de que sea de la máquina B)
*   $P(D|A) = 0.02$ (Probabilidad de defecto dado que es de A)
*   $P(D|B) = 0.05$ (Probabilidad de defecto dado que es de B)

$$P(D) = (P(A) \times P(D|A)) + (P(B) \times P(D|B))$$$$P(D) = (0.60 \times 0.02) + (0.40 \times 0.05)$$$$P(D) = 0.012 + 0.020 = \mathbf{0.032}$$

$$P(B|D) = \frac{P(B) \times P(D|B)}{P(D)}$$

$$P(B|D) = \frac{0.40 \times 0.05}{0.032}$$$$P(B|D) = \frac{0.020}{0.032}$$$$P(B|D) = \mathbf{0.625}$$

```python
import pandas as pd
import numpy as np


# Establece una semilla para reproducibilidad
np.random.seed(42)


# 1. Crea una lista de matrículas de estudiantes
student_ids = [f'{i:06d}' for i in range(172120, 172120 + 20)]


# 2. Asigna 10 estudiantes al grupo 'Café' y 10 al grupo 'Agua'
groups = ['Café'] * 10 + ['Agua'] * 10
np.random.shuffle(groups)


# 3. Simula calificacionen de examenes para el grupo café
# Mean=70, Std=5, asegurate de que los valores estén entre (50-100)
coffee_scores = np.random.normal(loc=70, scale=5, size=10).round(0)
coffee_scores = np.clip(coffee_scores, 50, 100).astype(int)


# 4. Simula calificaciones de examen para el grupo de 'Agua'
# Mean=75, Std=5, asegurate de que los valores estén entre (50-100)
water_scores = np.random.normal(loc=75, scale=5, size=10).round(0)
water_scores = np.clip(water_scores, 50, 100).astype(int)


# Combina los valores basado en el grupo
exam_scores = []
coffee_idx = 0
water_idx = 0
for group in groups:
    if group == 'Café':
        exam_scores.append(coffee_scores[coffee_idx])
        coffee_idx += 1
    else:
        exam_scores.append(water_scores[water_idx])
        water_idx += 1


# 5. Combina en un DataFrame de Pandas
df_students = pd.DataFrame({
    'Matrícula': student_ids,
    'Grupo': groups,
    'Calificaciones': exam_scores
})


print("Datos simulados:")
print(df_students.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('munsell-array-fixed-choice.csv')

idiomas = {
    'Inglés': 'english_color',
    'Español': 'spanish_color',
    'Tsimane': 'tsimane_color'
}

# Creo la figura con 3 subgráficos (1 fila, 3 columnas)
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for ax, (titulo, columna) in zip(axes, idiomas.items()):
    # Calculo el porcentaje de aparición de cada color
    porcentajes = df[columna].value_counts(normalize=True) * 100

    # Obtengo un color representativo para cada categoría
    hex_representativo = df.groupby(columna)['tile_hex'].agg(lambda x: x.mode()[0])
    colores_barras = [hex_representativo[color] for color in porcentajes.index]

    # Creo la grafica de barras
    porcentajes.plot(
        kind='barh',
        ax=ax,
        color=colores_barras,
        edgecolor='black'
    )

    # Inivierto el eje Y para que el porcentaje más alto quede arriba
    ax.invert_yaxis()

    # Doy formato al gráfico
    ax.set_title(f'Porcentaje de colores en {titulo}', fontsize=14, fontweight='bold')
    ax.set_xlabel('Porcentaje (%)', fontsize=12)  # El porcentaje está en el eje X
    ax.set_ylabel('Términos de color', fontsize=12) # Los nombres en el eje Y

    # Añadir una cuadrícula vertical de fondo para facilitar la lectura de los porcentajes
    ax.grid(axis='x', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('grafica_comparativa_horizontal.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('munsell-array-fixed-choice.csv')

counts = df.groupby(['english_color', 'spanish_color']).size().reset_index(name='frecuencia')

plt.figure(figsize=(8, 6))
plt.scatter(counts['english_color'], counts['spanish_color'], s=counts['frecuencia']*20, alpha=0.7)

plt.xlabel('Colores en Inglés')
plt.ylabel('Colores en Español')
plt.title('Correlación: Inglés vs Español')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

counts_et = df.groupby(['english_color', 'tsimane_color']).size().reset_index(name='frecuencia')

plt.figure(figsize=(8, 6))
plt.scatter(counts_et['english_color'], counts_et['tsimane_color'], s=counts_et['frecuencia']*20, alpha=0.7, color='orange')

plt.xlabel('Colores en Inglés')
plt.ylabel('Colores en Tsimane')
plt.title('Correlación: Inglés vs Tsimane')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

counts_st = df.groupby(['spanish_color', 'tsimane_color']).size().reset_index(name='frecuencia')

plt.figure(figsize=(8, 6))
plt.scatter(counts_st['spanish_color'], counts_st['tsimane_color'], s=counts_st['frecuencia']*20, alpha=0.7, color='green')

plt.xlabel('Colores en Español')
plt.ylabel('Colores en Tsimane')
plt.title('Correlación: Español vs Tsimane')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#Seleccione la opción que une los dataframes english_df y spanish_df y regresa los colores que existen en los dos dataframes.
import pandas as pd

df = pd.read_csv('munsell-array-fixed-choice.csv')

english_df = df['english_color'].value_counts().reset_index()
english_df.columns = ['english_color', 'count_en']

spanish_df = df['spanish_color'].value_counts().reset_index()
spanish_df.columns = ['spanish_color', 'count_es']

corr_ing_esp = pd.merge(english_df, spanish_df, left_on='english_color', right_on='spanish_color', how='inner')

print(corr_ing_esp.head())
```
