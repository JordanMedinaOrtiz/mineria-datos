####Conceptos

La minería de datos (o data mining) es el proceso de analizar grandes volúmenes de datos para descubrir patrones, anomalías y correlaciones ocultas, transformando información cruda en conocimiento valioso para la toma de decisiones estratégicas, la predicción de tendencias y la optimización de procesos en diversas industrias.
En otras palabras la minería de datos es inteligencia aplicada a la información. Consiste en usar programas informáticos avanzados para revisar montañas de datos y encontrar "pistas" que un ser humano jamás vería por su cuenta. Su objetivo no es solo saber qué pasó ayer, sino adivinar qué pasará mañana (como qué comprará un cliente o si una transacción es falsa) para tomar decisiones más inteligentes y rentables.

La ciencia de datos es un campo multidisciplinario que usa métodos científicos, procesos, algoritmos y sistemas para extraer conocimiento e insights valiosos de grandes volúmenes de datos (Big Data), combinando estadística, matemáticas, inteligencia artificial (IA) e informática para analizar, predecir y tomar decisiones informadas, abarcando desde la obtención hasta la aplicación de la información.
En otras palabras ña ciencia de datos es el ecosistema completo del manejo de información. No es solo una técnica, es una disciplina híbrida que mezcla tres mundos:
Matemáticas y Estadística: Para modelar la realidad.
Programación y Tecnología: Para procesar lo masivo (Big Data).
Conocimiento del Negocio: Para que los resultados tengan sentido útil.
Su misión abarca todo el ciclo de vida del dato: desde que nace (captura), se limpia, se analiza, hasta que se convierte en un producto o una decisión ejecutiva.

```python
# Sirve para evaluar el desempeño de una expresión en cuanto a tiempo
%timeit L = [n ** 2 for n in range(1000)]
```

```python
# Con doble porcentaje podemos evaluar el desempeño
# de en cuanto tiempo de más de una línea de código
%%timeit
L = []
for n in range(10000):
  L.append(n ** 2)
```
